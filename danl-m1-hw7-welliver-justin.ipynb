# Take Home Exam Q1 - Find the top 5 beer markets in terms of the number of households that purchased beer. Find the top 5 beer markets in terms of the amount of total beer consumption.

```python
top_beer_hh = beer_mkts.agg(hh_count = ('hh', 'count')) # Count the number of households for each market
top_beer_hh_sorted = top_beer_hh.sort_values(by='hh_count', ascending=False) # Sort the DF by hh column in descending order
top_5 = top_beer_hh_sorted.head(5) # Find the top 5 beer markets
top_5
```

```python
top_beer_consumption = beer_mkts.agg(beerfl_sum = ('beer_floz', 'sum')) # Add the volume of beer for each market
top_beer_consumption_sorted = top_beer_consumption.sort_values(by='beerfl_sum', ascending=False) # Sort the DF by beer_floz column in descending order
top_consumption = top_beer_consumption_sorted.head(5) # Find the top 5 beer markets
top_consumption
```

```python
import pandas as pd
beer_mkts = pd.read_csv('https://bcdanl.github.io/data/beer_markets.csv')
beer_mkts

bud_light_buyers = beer_mkts[beer_mkts['brand'] == 'BUD LIGHT']['hh'].unique() # Identify households that purchased BUD LIGHT at least once
only_bud_light_buyers = beer_mkts.groupby('hh').filter(lambda x: all(x['brand'] == 'BUD LIGHT'))['hh'].unique() # Identify households that purchased only BUD LIGHT
fraction_only_bud_light = len(only_bud_light_buyers) / len(bud_light_buyers) # Calculate the fraction of households that purchased only BUD LIGHT out of households that purchased BUD LIGHT at least once
fraction_only_bud_light
```

# Take Home Exam Q3 - Summarize the mean vale of revenue for each group of search_stays_on and for each date.

```python
import pandas as pd
import numpy as np
paidsearch = pd.read_csv('https://bcdanl.github.io/data/paidsearch.csv')
paidsearch['date'] = pd.to_datetime(paidsearch['date']) # Convert date column to datetime
mean_revenue_by_group = paidsearch.groupby(['date', 'search_stays_on'])['revenue'].mean() # Group by date and search_stays_on and calculate the mean revenue
mean_revenue_by_group
```

# Take Home Exam Q4 - Calculate the log difference between mean revenues in each group of search_stays_on.

```python
mean_revenue = paidsearch.groupby(['date', 'search_stays_on'])['revenue'].mean().reset_index() # Calculate mean revenue for each group of search_stays_on and for each date
pivot_mean_revenue = mean_revenue.pivot(index='date', columns='search_stays_on', values='revenue') # Pivot the DF to have date as index and each group of search_stays_on as a separate column
log_diff_mean_revenue = pivot_mean_revenue.diff(axis=1).apply(lambda x: np.log(x)) # Calculate the log difference between mean revenues in each group
log_diff_mean_revenue
```
