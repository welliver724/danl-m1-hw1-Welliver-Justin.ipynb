# HW5 Q1 - What percentage of the values is missing for each variable?

```python
percent_missing = fao.isna().mean().round(4)*100 # Find the average of the values missing for each variable
percent_missing
```

# HW5 Q2 - Fill missing values in the gdp_per_capita variable with the mean value of that variable.

```python
average = fao['gdp_per_capita'].mean().round(1) # Find the mean of the variable
average # Display the mean of the variable
```

```python
fao['gdp_per_capita'].fillna(value=average, inplace=True) # Replace missing values with the found mean value in the variable
fao
```

# HW5 Q3 - Drop observations where drinking_water or sanitation_service information is missing.

```python
fao = pd.read_csv("https://bcdanl.github.io/data/fao_stat.csv",
                  encoding = 'ISO-8859-1')
```

```python
fao.count() # Find the total number of observations for each variable in the DF before removing observations
```

```python
drop = fao.dropna(subset = ["drinking_water", "sanitation_service"], how = 'all') # Remove observations that have a missing value in the two variables
drop # Display DF
```

```python
drop.count() # Display updated obversation counts after removing observations that have a missing value
```

# HW5 Q4 What is the average drinking_water access percentage for each Area grouped by SSA status?

```python
(
  fao
  .groupby('Area')[['drinking_water', 'SSA']]
  .mean()
)
```

# HW5 Q5 - Calculate the mean sanitation_service percentage for each combination of SSA status and Year.

```python
(
  fao
  .groupby('SSA')[['sanitation_service']]
  .mean()
) # Mean sanitation_service percentage by SSA
```

```python
(
  fao
  .groupby('Year')[['sanitation_service']]
  .mean()
) # Mean Sanitation_service by Year
```

# HW5 Q6 - For each year, find the 5 worst countries in terms of drinking_water.

```python
(
    fao
    .groupby('Area')[['drinking_water']]
    .mean()
    .sort_values(['drinking_water'])
    .head(5)
) # Grouped the DF by area and drinking water. Found the average drinking water by each area. Sorted the values and listed the top 5 worst.
```
