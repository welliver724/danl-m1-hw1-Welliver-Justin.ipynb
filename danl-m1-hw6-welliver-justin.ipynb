```python
import pandas as pd

path = 'https://bcdanl.github.io/data/climate_finance.csv'
climate_finance = pd.read_csv(path)
```

# HW6 Q1 - For each party, calculate the total funding contributions that were disbursed or provided for mitigation projects for each year.

```python
contributions = climate_finance.groupby(['Year', 'Party'])['Contribution'].sum().reset_index() # Group DF by Year and Party, then summing the funding contributions
contributions
```

# HW6 Q2 - How many parties have provided positive funding contributions to other countries or regions for their adaptation projects for every single year from 2011 to 2018?

```python
years = climate_finance[climate_finance['Year'].between(2011, 2018)] # Filter DF to show years 2011 to 2018
sum = years.groupby(['Year', 'Party']).sum().reset_index() # Group DF by Year and Party, then sum the funding contributions
positive_funding = sum[sum['Contribution'] > 0].groupby('Year').size() # Count the number of parties with positive funding contributions for each year
positive_funding
```

# HW6 Q3 - What are the minimum, first quartile, median, thrid quartile, maximum, mean, and standard deviation of Close and Volume for each company?

```python
path = 'https://bcdanl.github.io/data/stock_2019_2024.csv'
stock = pd.read_csv(path)
stock
```

```python
stock2 = stock.groupby('company').agg({
    'Close': ['min', lambda x: x.quantile(0.25), 'median', lambda x: x.quantile(0.75), 'max', 'mean', 'std'],
    'Volume': ['min', lambda x: x.quantile(0.25), 'median', lambda x: x.quantile(0.75), 'max', 'mean', 'std']
}) # Group by Company and calculate statistics

stock2.columns = ['Close Min', 'Close 1st Quartile', 'Close Median', 'Close 3rd Quartile', 'Close Max', 'Close Mean', 'Close Std',
                      'Volume Min', 'Volume 1st Quartile', 'Volume Median', 'Volume 3rd Quartile', 'Volume Max', 'Volume Mean', 'Volume Std'] # Rename the columns for clarity
stock2
```

# HW6 Q4 - Find the 10 largest values for Volume for each company using a lambda function.

```python
largest_values = lambda x: x.nlargest(10) # lambda Function to get the 10 largest values for Volume for each company
top_volume = stock.groupby('company')['Volume'].apply(largest_values) # apply the function
top_volume
```

# HW6 Q5 - Calculate the Z-score of Close for each company on each Date t.

```python
def calculate_zscore(x):
    return (x - x.mean()) / x.std() # Define a function to calculate Z-score

stock['Zscore'] = stock.groupby(['company', 'Date'])['Close'].transform(calculate_zscore) # Calculate Z-score of Close for each company on each Date
stock
```

# HW6 Q6 - Use the transform() method on the stock data to represent all the values of Open, High, Low, Close, Adj Close, and Volume in terms of the first date in the data.

```python
stock['Date'] = pd.to_datetime(stock['Date']) # Convert Date column to datetime
stock[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] = stock.groupby('Date')[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].transform(lambda x: x / x.iloc[0]) # Group by Date and transform the values
stock
```
