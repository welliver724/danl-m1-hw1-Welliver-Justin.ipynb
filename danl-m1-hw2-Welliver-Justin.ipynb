```python
# Import pandas Library

import pandas as pd
```

```python
# Import nhl DataFrame

nhl = pd.read_csv('https://bcdanl.github.io/data/NHL-1617.csv',
                  parse_dates = ["Born"])
nhl
```

```python
# HW2 Q1 - How Many Players Had At Least Some Ice Time?

nhl[['TOI']].count()
```

The total number of players who had some ice time is 888. I found this using .count() to count the number of non-missing values in a Series/DataFrame.

```python
# HW2 Q2 - Who is the top scorer in terms of goals?

nhl.nlargest(1, 'G')
```

The top goal scorer in the DataFram is Cole Harbour. I'm sure there are many ways to get the top scorer in terms of goals, but I used the nlargest function. nlargest is good to get the first n observation ordered by a variable in descending order.

```python
# HW2 Q3 - How can we count the number of NHL players for each country?

nhl[['Cntry']].value_counts()
```

.value_counts() counts the number of occurances of each unique value in a Series/DataFrame. I used .value_counts() to determine the number of NHL players that play for each country.

```python
# HW2 Q4 - How many unique countries are in nhl.csv?

nhl[['Cntry']].nunique()
```

.nunique() counts the number of unique values in each variable in a Series/DataFrame. I used .nunique() to determine that there are 19 unique countries in the DataFrame.

```python
# HW2 Q5 - Which three nationalities have the highest number of players?

df = nhl['Nat'].value_counts()
df.head(3)
```

To find which 3 nationalities have the highest number of players, I first needed to count the number of occurances for each nationality in the Series/DataFrame. I did this by using .value_counts() on the 'Nat' column. I then used .head(3) to show only the top 3 largest values. I found that Canada, USA, and Sweden are the 3 top nationalities in the DataFrame.
