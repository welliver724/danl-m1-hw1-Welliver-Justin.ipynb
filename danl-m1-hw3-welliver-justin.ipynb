```python
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
nyc_payroll = pd.read_csv('https://bcdanl.github.io/data/nyc_payroll_2022_2023.csv',
                          parse_dates = ["Agency Start Date"])

nyc_payroll

# HW3 Q6 - Select employees who have "OT Hours" greater than 0, calculate their "OT Rate", and then find the employee with the highest "OT Rate". Display their full name and OT Rate

# Boolean Series of whether or not 'OT Hours' is greater than 0.
pos_OT_hours = nyc_payroll['OT Hours'] > 0

# Keeping only the observations with pos_OT_hours == True
with_ot = nyc_payroll[ pos_OT_hours ]

# Calculating OT Rate
with_ot["OT Rate"] = with_ot["Total OT Paid"] / with_ot["OT Hours"]

#Find Highest OT Rate
with_ot.sort_values(["OT Rate"],
                    ascending = False)

# Display their full name and OT Rate
nyc_payroll7 = with_ot[['First Name','Last Name' , 'OT Rate']]

# Find highest OT Rate
nyc_payroll7.nlargest(1, 'OT Rate')

```

The running code is now displaying the employee with the highest OT Rate by full name. The individuals name is Daniel Debonis and his OT Rate is 174.116.

```python
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
nyc_payroll = pd.read_csv('https://bcdanl.github.io/data/nyc_payroll_2022_2023.csv',
                          parse_dates = ["Agency Start Date"])

nyc_payroll

# Create "Total Compensation" column
nyc_payroll['Total Compensation'] = nyc_payroll['Base Salary'] + nyc_payroll['Total OT Paid']

# Sort DF by "Work Location Borough" alphabetically
nyc_payroll10 = nyc_payroll[['First Name', 'Last Name', 'Base Salary', 'Total OT Paid', 'Work Location Borough', 'Total Compensation' ]]
nyc_payroll11 = nyc_payroll10.sort_values(["Work Location Borough", "Total Compensation"],
                    ascending = [True, False])
nyc_payroll11
```
